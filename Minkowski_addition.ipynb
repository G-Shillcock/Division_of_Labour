```python
# Cartesian

import matplotlib.pyplot as plt
import numpy as np
import itertools as it


def B(A):
    return (1-A**(1/a))**b


def brute():
    Asum = []
    Bsum = []

    for As in it.product(np.linspace(0,1,20),repeat=n):

        Asum.append(sum(As))
        Bsum.append(sum( (1-A**(1/a))**b for A in As) )

    return Asum, Bsum


def scalloped():

    Asum = np.linspace(0,n,1000)[1:-1]

    Bsum = np.floor(n-Asum) + B(Asum % 1)

    return Asum, Bsum


def scaled():
    Asum = np.arange(0,n,n/1000)
    Bsum = n*B(Asum/n)

    return Asum, Bsum


a = 1.5
b = 1.3
n = 5

Av = np.linspace(0,1)
Bval = (1-Av**(1/a))**b

plt.figure(figsize=(4,4))
plt.plot(Av,Bval)

Asum, Bsum = brute()
plt.plot(Asum, Bsum,'b.')

Asum, Bsum = scalloped()

plt.plot(Asum, Bsum)

Asum, Bsum = scaled()
plt.plot(Asum, Bsum);
```

```python
# Parametric

import matplotlib.pyplot as plt
import numpy as np
import itertools as it

a = 1.5
b = a
n = 4

xs = []
ys = []

Av = np.linspace(0,1)
Bval = (1-Av**(1/a))**b

plt.figure(figsize=(4,4))
plt.plot(Av,Bval)

for ts in it.product(np.linspace(0,1,10),repeat=n):

    xs.append(sum(t**a for t in ts))
    ys.append(sum((1-t)**b for t in ts))

plt.plot(xs,ys,'b.')
print(len(xs),len(set(xs))) # alot of redundancy

xs = []
ys = []
for ts in it.product(np.linspace(0,1,10),repeat=n):

    if ts.count(0)==n//2 or ts.count(1)==n//2:
        xs.append(sum(t**a for t in ts))
        ys.append(sum((1-t)**b for t in ts))

plt.plot(xs,ys,'rx')
print(len(xs),len(set(xs))) # alot of redundancy
```

```python
def B(A):
    return (1-A**(1/a))**b


def brute_product():
    Aprod = []
    Bprod = []

    for As in it.product(np.linspace(0,1,20),repeat=n):

        Aprod.append( np.prod(As) )
        Bprod.append( np.prod( [B(A) for A in As] ) )

    return Aprod, Bprod

# a = 0.1
# b = 0.5
n = 3

fig, ax = plt.subplots(3,3,figsize=(4,4))
for i, a in enumerate([.1,.5,2]):
    for j, b in enumerate([.1,.5,2]):

        Aprod, Bprod = brute_product()
        ax[i,j].plot(Aprod, Bprod,'b.')
        ax[i,j].set_title(f'{a,b}')
plt.tight_layout()
```
